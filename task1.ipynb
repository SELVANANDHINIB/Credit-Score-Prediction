```python
b=df.drop(columns='CUST_ID')
```

```python
X_train, X_test, y_train, y_test = train_test_split(a.drop('CREDIT_SCORE', axis=1), a['CREDIT_SCORE'], test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
d = RandomForestClassifier(n_estimators=100, random_state=42)
d.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
pre=d.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
# Example conversion: if the values are probabilities, convert to binary labels
threshold = 0.5
pred_labels = np.where(pre >= threshold, 1, 0)
y_test_labels = np.where(y_test >= threshold, 1, 0)

print("Accuracy:", accuracy_score(pred_labels, y_test_labels))
```
